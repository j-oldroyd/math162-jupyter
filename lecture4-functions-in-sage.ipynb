# Defining Functions in Sage

- The **linear approximation** of a differentiable function $f(x)$ at a point $x = a$ is given by
$$ L(x) = f(a) + f'(a)(x - a).$$

**Example 1**: Find the linear approximation $L(x)$ to $f(x) = e^x$ at $x = 0$ and plot both $f$ and $L$.

````python
%display latex
# enter code for Example 1 here
f(x) = exp(x)
g(x) = derivative(f(x), x)
a = 0
L(x) = f(a) + g(a) * (x - a)

p1 = plot(f(x)) # plot of e^x
p2 = plot(L(x), color = 'red') # plot of L(x)
P = p1 + p2
P.show()
````

- Now suppose we wanted to find the linear approximation at another point, or for another function. When you find yourself repeating code, it's often useful to define a *function* that lets you reuse the code easily. In Python and Sage, functions can be defined as follows:
```python
def function(in1, in2, ...): # note the colon!
    # fill in code that you want to reuse
    # should depend on what your inputs are
    return out1, out2, ... # value represents the function's output
```
Here, `in1`, `in2`, and so on are **input arguments**. Likewise, `out1`, `out2`, and so on are **output arguments**. When defining functions like this, **indentation is important**.

- For example, if we want to create approximations using the formula
$$L(x) = f(a) + f'(a)(x-a)$$
for more functions, we could define a `lin_approx` function in Sage that lets us reuse our code from before. Since the formula depends on the function $f$ and the number $a$, we make those the input arguments of our function. Likewise, we make $L(x)$ the output as that's what we want this function to produce.

````python
def lin_approx(f, a):
    """
    Computes linear approximation of f(x) at x = a.
    """
    g(x) = f(x).derivative()
    L(x) = f(a) + g(a) * (x - a) # f(a) + f'(a)(x - a)
    return L(x)
````

**Example 2**: Use the new `lin_approx` function to find the linear approximation of $\ln x$ at $x = 1$.

````python
# enter code for Example 2 here
lin_approx(log, 1)
````

- Functions in Python and Sage can include documentation to explain their use. This can be accessed in the notebook by using the format `function_name?`. For example, to see the documentation of the `lin_approx` function defined above, we can do the following:

````python
lin_approx?
````

**Example 3**: The **quadratic approximation** to a differentiable function $f(x)$ at $x = a$ is defined by
$$ Q(x) = f(a) + f'(a)(x - a) + \frac{1}{2}f''(a)(x - a)^{2}.$$
In the code cell below, find the quadratic approximation to $f(x) = e^x$ at $x = 0$ and plot $e^x$, its linear approximation $L(x)$ and its quadratic approximation $Q(x)$ at $x = 0$.

````python
#enter code for Example 3 here
f(x) = exp(x)
g(x) = derivative(f(x),x)
h(x) = derivative(g(x),x)
a = 0
Q(x) = f(a) + g(a) * (x - a) + (h(a) / 2) * (x - a)^2

p1 = plot(f(x))
p2 = plot(lin_approx(exp, 0), color = 'red')
p3 = plot(Q(x), color = 'green')
P = p1 + p2 + p3
P.show()
````
